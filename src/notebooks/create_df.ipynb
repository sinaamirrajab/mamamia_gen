```python
from sklearn.model_selection import train_test_split
import pandas as pd
import os
def get_data_files_df(df_dir):
    df = pd.read_csv(df_dir)
    # Create train, validation and test dataset.
    df['age'] = pd.to_numeric(df['age'].fillna(df['age'].median()), errors='coerce')
    df['her2'] = pd.to_numeric(df['her2'].fillna(df['her2'].median()), errors='coerce')
    df['tumor_subtype'] = df['tumor_subtype'].fillna('Unknown')

    unique_df = df.drop_duplicates(subset='patient_id', keep='first')

    unique_train_df, unique_validation_df = train_test_split(
            unique_df,
            train_size=0.80,
            test_size=0.20,
            stratify=unique_df['pcr'],
            random_state=0,
            shuffle=True,)
    # Filter the original DataFrame based on train_df patient_id and keep the first two image_id entries
    train_df = df[df['patient_id'].isin(unique_train_df['patient_id'])]
    # train_df = train_df.sort_values(by=['patient_id', 'image_id'])
    train_df = train_df.sort_values(by=['patient_id'])
    # train_df = train_df.groupby('patient_id').head(images_per_patient).reset_index(drop=True)

    validation_df = df[df['patient_id'].isin(unique_validation_df['patient_id'])]
    # validation_df = validation_df.sort_values(by=['patient_id', 'image_id'])
    validation_df = validation_df.sort_values(by=['patient_id'])
    # validation_df = validation_df.groupby('patient_id').head(images_per_patient).reset_index(drop=True) # taking the first two aquisitions of each patient

    #  Define the essential fields
#     essential_fields = ["patient_id","image_id", "image_path", "pcr", "age", "tumor_subtype", "her2", "seg_path"]
    essential_fields = ["patient_id","pcr", "age", "tumor_subtype", "her2"]


    # Filter data to include only essential fields
    filtered_train_ds = [{key: item[key] for key in essential_fields} for item in train_df.to_dict("records")]
    filtered_validation_ds = [{key: item[key] for key in essential_fields} for item in validation_df.to_dict("records")]
    return filtered_train_ds, filtered_validation_ds


def get_train_df(df_dir):
    train_df = pd.read_csv(df_dir)

    essential_fields = ["patient_id","image_id", "image_path", "pcr", "age", "tumor_subtype", "her2", "seg_path"]

    # Filter data to include only essential fields
    filtered_train_ds = [{key: item[key] for key in essential_fields} for item in train_df.to_dict("records")]
    return filtered_train_ds
```

```python


def get_unique_patient_ids(data):
    seen = set()
    unique_ids = []
    for item in data:
        pid = item['patient_id']
        if pid not in seen:
            seen.add(pid)
            unique_ids.append(pid)
    return unique_ids

# Get unique patient_ids in order
unique_train_ids = get_unique_patient_ids(train_ds)
unique_val_ids = get_unique_patient_ids(train_ds_list)

# Find unique-to-train and unique-to-validation
unique_in_train_only = [pid for pid in unique_train_ids if pid not in unique_val_ids]
unique_in_val_only = [pid for pid in unique_val_ids if pid not in unique_train_ids]

print("Unique patient_ids in train only:", unique_in_train_only)
print("Unique patient_ids in validation only:", unique_in_val_only)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
plot_phaese = 'valid'

df_dir = r"/projects/0/prjs1204/projects/spadebreast2025/src/spadebreast/dataset/combined_df/stratified_96_train_valid_df_seg.csv"
# df_dir = r"/projects/0/prjs1204/projects/spadebreast2025/src/spadebreast/dataset/combined_df/stratified_96_test_df.csv"
df = pd.read_csv(df_dir)

unique_df = df.drop_duplicates(subset='patient_id', keep='first')
unique_train_df, unique_validation_df = train_test_split(
                unique_df,
                train_size=0.80,
                test_size=0.20,
                stratify=unique_df['pcr'],
                random_state=0,
                shuffle=True,)

 # Filter the original DataFrame based on train_df patient_id and keep the first two image_id entries
train_df = df[df['patient_id'].isin(unique_train_df['patient_id'])]
# train_df = train_df.sort_values(by=['patient_id', 'image_id'])
train_df = train_df.sort_values(by=['patient_id'])
# train_df = train_df.groupby('patient_id').head(images_per_patient).reset_index(drop=True)

validation_df = df[df['patient_id'].isin(unique_validation_df['patient_id'])]
# validation_df = validation_df.sort_values(by=['patient_id', 'image_id'])
validation_df = validation_df.sort_values(by=['patient_id'])
# Group by tumor_subtype and pcr, then count unique patient_id values
if plot_phaese == 'train':
    unique_patient_counts = train_df.groupby(['tumor_subtype', 'pcr'])['patient_id'].nunique()
elif plot_phaese == 'valid':
    unique_patient_counts = validation_df.groupby(['tumor_subtype', 'pcr'])['patient_id'].nunique()
else:
    raise ValueError("Invalid plot phase. Use 'train' or 'valid'.")
# Convert to DataFrame for plotting
plot_data = unique_patient_counts.unstack()

# Plot
ax = plot_data.plot(kind='bar', figsize=(10, 6))

plt.title('Number of Unique Patients per Tumor Subtype and pCR Status')
plt.xlabel('Tumor Subtype')
plt.ylabel('Number of Unique Patients')
plt.xticks(rotation=45)
plt.legend(title='pCR')

# Annotate bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2, height), 
                    ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import os
# Step 1: Group and get unique patient_ids for each (tumor_subtype, pcr)
# df  = train_df
df = validation_df
grouped = df.groupby(['tumor_subtype', 'pcr'])['patient_id'].unique()

# Step 2: Loop through each category and filter original DataFrame
filtered_dfs = {}  # Dictionary to store category -> DataFrame
# out_dir = r'/projects/0/prjs1204/projects/spadebreast2025/src/spadebreast/dataset/combined_df_syn/train_df'
out_dir = r'/projects/0/prjs1204/projects/spadebreast2025/src/spadebreast/dataset/combined_df_syn/validation_df'
os.makedirs(out_dir, exist_ok=True)
for (subtype, pcr_status), patient_ids in grouped.items():
    # Filter original DataFrame where patient_id is in the group
    filtered_df = df[df['patient_id'].isin(patient_ids)]
    print(f"size of {subtype} pCR {pcr_status}: {len(filtered_df)}")
    
    # Optional: Reset index
    filtered_df = filtered_df.reset_index(drop=True)
    
    # Save to dictionary with a clean key name
    key_name = f"{subtype}_pcr{int(pcr_status)}"
    filtered_dfs[key_name] = filtered_df

    # Step 3: Save each DataFrame to CSV (or pickle, etc.)
    filtered_df.to_csv(os.path.join(out_dir, f"{key_name}.csv"), index=False)
    filtered_df.head(100)

print("All filtered DataFrames saved by tumor subtype and pCR status.")
```
